```python
df = df[df["PropertyType"] == "Residential"]

df.shape
```

```python
df_null_cnt = (round(df.isnull().sum() / df.shape[0] * 100, 2)).reset_index()
df_null_cnt.columns = ["ColName", "NullPct"]
df_null_cnt.head()
```

```python
df_null_cnt["isSparse"] = df_null_cnt["NullPct"] >= 90.0
df_null_cnt
```

```python
date_cols = [col for col in df.columns if "date" in col.lower()]
date_cols
```
